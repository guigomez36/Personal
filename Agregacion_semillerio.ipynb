```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
df=None
#def myfunction(data):
#  print(data)

dir = '/content/drive/MyDrive/Colab Notebooks/DMEyF/competencia_FINAL/agrupacion_semillas/in'

for file in os.listdir(dir):
  print(file)
  if file.endswith(".csv"):
    new=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DMEyF/competencia_FINAL/agrupacion_semillas/in/' + file, sep='\t')
    new=new.sort_values(by='numero_de_cliente', ascending=True)
    if df is not None:
      df=pd.concat([df,new[['prob']]], axis=1)
    else:
      df=new[['numero_de_cliente','foto_mes','prob']]
    #myfunction(file)
```

```python
df.head()
```

```python
df['mean'] = df.iloc[:, 2:(df.shape[1]-2)].mean(axis=1)
```

```python
df.head()
```

```python
df_final=df[['numero_de_cliente', 'mean']].sort_values(by='mean', ascending=False)
df_final=df_final.reset_index()
```

```python
df_final.head()
```

```python
df_final['Predicted_12000']=0
for i in range(12000):
  df_final['Predicted_12000'][i]=1
```

```python
df_final[['numero_de_cliente','Predicted_12000']].to_csv(dir + '/agrupado_12000.csv', index=False)
```
